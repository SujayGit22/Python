```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0.,5,0.2)

#red cross, blue square. green triangle
plt.plot(t,t,'r--',t,t**2,'bs',t,t**3,'g^')
plt.show()
```

```python
''' Working with multiple axis'''

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure(1)
plt.subplot(211)  # 2- subplots, 1- first figure,1-first subplot
plt.plot(t1,f(t1),'bo',t2,f(t2),'k')

plt.subplot(212)  # 2-subplots, 1-first figure, 1- first subplot
plt.plot(t2,np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
# Working with Text
# All of the text() commands return an matplotlib.text.Text instance
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='g', alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
